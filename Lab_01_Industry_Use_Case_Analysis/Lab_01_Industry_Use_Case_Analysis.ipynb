# Lab 01 - Industry Use Case Analysis

## Storytelling Goal

In this lab, we do not start with code. We start with a business story:

- What problem exists?
- Who is affected?
- What data could help?
- What AI decision or recommendation is needed?
- What action should a human take after seeing the AI output?
