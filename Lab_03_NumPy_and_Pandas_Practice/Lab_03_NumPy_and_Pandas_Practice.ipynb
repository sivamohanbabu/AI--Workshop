## Data Story

We are analyzing salary data to understand how experience relates to salary.

Business question:

**How does salary grow as years of experience increase?**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset/Salary_dataset.csv")
df = df.drop(columns=["Unnamed: 0"], errors="ignore")
df.head()
```

## 1. Basic Inspection

```python
print("Rows and columns:", df.shape)
display(df.info())
display(df.describe().T)
```

## 2. Pandas Practice

Create useful business summaries.

```python
average_salary = df["Salary"].mean()
min_experience = df["YearsExperience"].min()
max_experience = df["YearsExperience"].max()

print("Average salary:", round(average_salary, 2))
print("Minimum experience:", min_experience)
print("Maximum experience:", max_experience)

df["Salary_Level"] = np.where(df["Salary"] >= average_salary, "Above Average", "Below Average")
df.head()
```

## 3. Visual Story

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["YearsExperience"], df["Salary"], s=80, alpha=0.75)
plt.title("Salary Increases with Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## 4. Evidence Table

```python
df.sort_values("YearsExperience").head(10)
```

## 5. Data Storytelling Conclusion

```python
correlation = df["YearsExperience"].corr(df["Salary"])
print(f"Experience and salary correlation: {correlation:.2f}")
print("Story conclusion: More experience is strongly associated with higher salary in this dataset.")
print("Business use: This can support compensation planning, but more features are needed for fair salary decisions.")
```
