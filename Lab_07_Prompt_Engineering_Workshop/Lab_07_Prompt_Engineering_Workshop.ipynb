# Lab 07 - Prompt Engineering Workshop

## Storytelling Goal

Prompt engineering is also storytelling. A strong prompt gives the AI:

1. Role
2. Context
3. Task
4. Format
5. Constraints
6. Example, if useful

## Weak Prompt

```python
weak_prompt = "Write about AI."
print(weak_prompt)
```

## Strong Prompt

```python
strong_prompt = '''
Act as an AI trainer teaching beginners.
Explain artificial intelligence in 5 bullet points.
Use examples from banking, education, and healthcare.
Avoid technical jargon.
End with one simple question for learners.
'''

print(strong_prompt)
```

## Prompt Builder Template

```python
role = "Act as a data analyst"
context = "The audience is beginner AI learners"
task = "Explain the result of a machine learning model"
format_required = "Use 4 bullet points"
constraints = "Use simple language and avoid jargon"

prompt = f'''
{role}.
Context: {context}.
Task: {task}.
Format: {format_required}.
Constraints: {constraints}.
'''

print(prompt)
```
